```python
import sys

import cv2 as cv
import matplotlib.pyplot as plt
```

```python
img = cv.imread("starry_night.jpg")
```

```python
plt.imshow(img)
img_rgb = cv.cvtColor(img, cv.COLOR_BGR2RGB)
plt.imshow(img_rgb)
```

```python
def bgr2rgb(a):
    a[:,:,[0,1,2]] = a[:,:,[2,1,0]]
    return a
```

```python
img_rgb2 = bgr2rgb(img)
plt.imshow(img_rgb2)
```

```python
def mmc (a):
    b = a.copy()
    c = a[130:250, 300:410,:]
    d = a[50:170, 620:730,:]
    b[130:250, 300:410,:] = d
    b[50:170, 620:730,:] = c 
    return b
```

```python
qqq  = mmc(img_rgb2)
plt.imshow(qqq)
```

```python
sss = img_rgb2.copy()
sss[:,:,[0]], sss[:,:,[1]], sss[:,:,[2]] = (sss[:,:,[0]]*1.2, sss[:,:,[1]]*0.95, sss[:,:,[2]]*0.6)
plt.imshow(sss)
```
