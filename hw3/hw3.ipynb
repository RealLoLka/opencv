```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
from math import sin, cos, pi
```

```python
img = cv.imread("messi5.jpg")
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
height, width = img.shape[:2]
plt.imshow(img)
```

```python
angle = 90
o = (width/2, height/2)


R = cv.getRotationMatrix2D(o, angle, 1)
rimg = cv.warpAffine(img, R, (width, height))
plt.imshow(rimg)
```

```python
img2 = cv.imread("apple.jpg")
plt.imshow(img2)
```

```python
height, width = img2.shape[:2]
center = (width/2,height/2)
angle = 180
M = cv.getRotationMatrix2D(center, angle, 1)
rimg2 = cv.warpAffine(img2, M, (width, height))
plt.imshow(rimg2)
```

```python
img3 = cv.imread("phone.png")
plt.imshow(img3)
```

```python
S = np.float32([
    [0.5, 0, 0],
    [0, 1, 0],
])
height, width = img3.shape[:2]
rimg3 = cv.warpAffine(img3, S, (width, height))
plt.imshow(rimg3)
```

```python
img4 = cv.imread("skull.png")
plt.imshow(img4)
```

```python
angle = 45
height, width = img3.shape[:2]
o = (width/2, height/2)

R = cv.getRotationMatrix2D(o, angle, 1)
rimg4 = cv.warpAffine(img4, R, (width, height))

S = np.float32([
    [3.5, 0, 0],
    [0, 1.5, 0],
])

M = S * R
rimg4 = cv.warpAffine(rimg4, M, (int(width * 2), height))

plt.imshow(rimg4)
```
