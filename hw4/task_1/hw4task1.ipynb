```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt
```

```python
resimg = cv.imread("railway_result.png")
plt.imshow(resimg)
```

```python
img = cv.imread("railway.jpeg")
plt.imshow(img)
```

```python
height, width = img.shape[:2]
pts1 = [
    [250, 650],
    [120, 800],
    [750, 650],
    [900, 800],
]
pts2 = [
    [0, 0],
    [0, height],
    [width, 0],
    [width, height]
]

img2 = img.copy()
for p1, p2 in zip(pts1, pts2):
    cv.circle(img2, center=p1, radius=20, color=(0, 0, 255), thickness=-1)
    cv.circle(img2, center=p2, radius=20, color=(0, 0, 255), thickness=-1)
    cv.arrowedLine(img2, p1, p2, color=(0, 0, 255), thickness=10) 
    
plt.imshow(img2, cmap='gray')
```

```python
M = cv.getPerspectiveTransform(np.float32(pts1), np.float32(pts2))
rimg = cv.warpPerspective(img, M, (width, height))
plt.imshow(rimg)
```
